## Separatively combine data for Camera1 and Camera2.
1. When processing Camera1, only use:
```python
camera_dir = 'occupancy_estimation/occupancy_estimation_camera1'

```
2. When processing Camera2, only use:
```python
camera_dir = 'occupancy_estimation/occupancy_estimation_camera2'
camera_dir_combined = 'occupancy_estimation/occupancy_estimation_camera2_combined'
```

````python
# # Camera1
# camera_dir = 'occupancy_estimation/occupancy_estimation_camera1'
# camera_dir_combined = 'occupancy_estimation/occupancy_estimation_camera1_combined'

# Camera2
camera_dir = './occupancy_estimation_camera2'
camera_dir_combined = './occupancy_estimation_camera2_combined'

camera_sub_dir = [sub_dir for sub_dir in os.listdir(camera_dir) if not sub_dir.startswith('.')]
camera_sub_dir.sort(reverse = False)
camera_sub_dir
````

````python
for sub_dir in camera_sub_dir:
    filepath = glob.glob(os.path.join(camera_dir, sub_dir,'*.csv'))
    filepath.sort()
    occupancy_combined = {}
    occupancy_time_point = {}
    
    # combine occupancy
    for file_path_temp in filepath:
        occupancy = pd.read_csv(file_path_temp)
        
        _, _, _, file = file_path_temp.split('/')
        info, info_openpose = file.split('_processed_')
        area = info_openpose.split('occupancy_')[1].split('_COCO')[0]
        time = info[8:24]
        behavior = info[25:][:-2]
        clip_num = info[25:][-1]
        
        # save end time of last clip
        occupancy_time_point[clip_num] = occupancy.iloc[-1]['Time']
        
        new_filename = '{}_processed_{}'.format(info[:-2], info_openpose)
        if new_filename not in occupancy_combined.keys():
            print('new_filename:{}'.format(new_filename))
            occupancy_combined[new_filename] = pd.DataFrame()
        
        for new_file in occupancy_combined.keys():
            if (area in new_file) and (time in new_file) and (behavior in new_file):
                # change time
                t = 0
                for i in range(int(clip_num)):
                    t += occupancy_time_point[str(i)]
                occupancy['Time'] += t
                occupancy_combined[new_file] = occupancy_combined[new_file].append(occupancy, ignore_index=True)
    
    # save combined occupancy as .csv file
    for combined_file in occupancy_combined.keys():
        camera_sub_dir_combined = os.path.join(camera_dir_combined, sub_dir)
        if not os.path.exists(camera_sub_dir_combined):
            os.makedirs(camera_sub_dir_combined)
        occupancy_combined[combined_file].to_csv(os.path.join(camera_sub_dir_combined,combined_file), encoding='utf-8', index=False)
````
